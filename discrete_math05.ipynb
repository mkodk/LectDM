```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return 2 * x + 1

x = np.linspace(-10, 10, 100)
y = f(x)

plt.plot(x,y,label = "f(x) = 2x + 1")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Example of function of graph")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def is_injective(f,domain):
  image = set()
  for x in domain :
    fx = f(x)
    if fx in image :
      return False
    image.add(fx)
  return True

domain = range(-10,11)
print("is it injective function?",is_injective(f,domain))
```

```python
def inverse_f(y):
  return(y-1)/2

x_values = np.linspace(-10,10,100)
y_values = inverse_f(x_values)

plt.plot(x_values,y_values,label = "inverse of f(x)",color = "red")
plt.xlabel('x')
plt.ylabel("f^{-1}(y)")
plt.title("inverse function graph")
plt.legend()
plt.grid(True)
plt.show()
```
